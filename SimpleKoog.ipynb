# Let's try Koog!

```kotlin
%useLatestDescriptors
%use coroutines
%use koog
```

```kotlin
val executor = simpleGoogleAIExecutor(System.getenv("GEMINI_API_KEY")!!)

runBlocking {
    AIAgent(
        executor = executor,
        llmModel = GoogleModels.Gemini2_5Flash,
        systemPrompt = "You are a helpful assistant powered by Koog."
    ).run("Hello, Koog!")
}
```

```kotlin
import ai.koog.agents.core.tools.annotations.Tool
import java.time.LocalDateTime
import java.util.TimeZone

@Tool
@LLMDescription("Get the current local date and time.")
fun currentTime(): String = LocalDateTime.now().toString()

val timeAgent = AIAgent(
    executor = executor,
    llmModel = GoogleModels.Gemini2_5Flash,
    toolRegistry = ToolRegistry { tool(::currentTime) },
    systemPrompt = "You should respond with human friendly text."
)
```

```kotlin
runBlocking { timeAgent.run("Hello! What's the time?") }
```
